```python
import asyncio
from pyppeteer import launch

async def screenshot():
    browser = await launch(
        handleSIGINT=False,
        handleSIGTERM=False,
        handleSIGHUP=False
    )
    print("hi")
    page = await browser.newPage()
    await page.goto('http://localhost:8989/app/?config=czXENrDfQIqeKZU-Mw2r7w', {
        'waitUntil': 'networkidle2',
    })
    await page.screenshot({'path': '/tmp/example.png'})
    await browser.close()
```

```python
await screenshot()
```
